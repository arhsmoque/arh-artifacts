```python
# @markdown #Inference Setup

from pathlib import Path
import glob
import os
import subprocess

# @param {type:"string"}
input_source = '/content/drive/MyDrive/00 FLASH SR INPUT'
# @param {type:"string"}
output = '/content/drive/MyDrive/00 FLASH SR OUTPUT'
# @param {type:"slider", min:2, max:10, step:1}
overlap = 2

input_files = []
if os.path.isdir(input_source):
    # If input_source is a directory, find all supported audio files within it
    for ext in ['*.wav', '*.mp3', '*.flac']: # Added flac based on user input
        input_files.extend(glob.glob(os.path.join(input_source, ext)))
elif os.path.isfile(input_source):
    # If input_source is a single file, add it to the list
    # Check if the single file has a supported extension
    if any(input_source.lower().endswith(ext) for ext in ['.wav', '.mp3', '.flac']):
         input_files.append(input_source)
    else:
        print(f"🔴 ERROR: Single input file {input_source} has an unsupported extension. Supported: .wav, .mp3, .flac")
else:
    print(f"🔴 ERROR: Invalid input source: {input_source}. It must be a directory or a single supported audio file.")

# Ensure the output directory exists
os.makedirs(output, exist_ok=True)
print(f"Ensured output directory exists: {output}")

print(f"\nFound {len(input_files)} input files to process.")
print("Input files:", input_files)
```

```python
#@markdown #Run Inference

from pathlib import Path
import os
import subprocess
import time

# Check if input_files list is populated from the previous cell
if not 'input_files' in locals() or not input_files:
    print("🔴 ERROR: No input files found or input_files list is empty. Please run the Inference Setup cell first.")
else:
    print(f"Starting inference for {len(input_files)} files...")

    flashsr_dir = "/content/FlashSR_Inference"
    # Ensure we are in the correct directory to run inference.py
    if os.path.exists(flashsr_dir) and os.path.isdir(flashsr_dir):
        %cd {flashsr_dir}
        print(f"Changed directory to {flashsr_dir}")
    else:
        print(f"🔴 ERROR: FlashSR_Inference directory not found at {flashsr_dir}. Please ensure the installation step ran successfully.")
        # If directory doesn't exist, we cannot run inference.py, so exit the process
        input_files = [] # Clear input_files to prevent loop from running

    # Ensure the output directory exists (redundant with setup cell, but good practice)
    os.makedirs(output, exist_ok=True)
    print(f"Ensured output directory exists: {output}")

    if input_files: # Check again if input_files is not empty after directory check
        total_files = len(input_files)
        for i, input_file in enumerate(input_files):
            # Construct the expected output filename with the "upscaled_" prefix and .flac extension
            input_filename = os.path.basename(input_file)
            name, ext = os.path.splitext(input_filename)
            # Construct the full path for the expected output file in the output directory
            expected_output_file_path = os.path.join(output, f"upscaled_{name}.flac") # Assuming output is always FLAC

            print(f"\n--- Processing file {i+1}/{total_files} ---")
            print(f"  Input file: {input_file}")
            print(f"  Expected output file: {expected_output_file_path}")

            # Check if the expected output file already exists
            if os.path.exists(expected_output_file_path):
                print(f"  Skipping: Expected output file already exists at {expected_output_file_path}")
                continue # Skip to the next file

            # Calculate progress percentage
            progress_percent = (i / total_files) * 100
            # Print the command that will be executed
            print(f"  Progress: ({progress_percent:.2f}%)")
            print(f"  Output Directory (passed to inference.py): {output}")


            try:
                # Run the inference command for the current file using subprocess for better control
                # Passing the output directory to the inference script, as it seems this is how it expects it
                print(f"  Executing command: python inference.py --input \"{input_file}\" --output \"{output}\" --overlap {overlap}")
                process = subprocess.Popen([
                    "python", "inference.py",
                    "--input", input_file,
                    "--output", output, # Pass the output directory
                    "--overlap", str(overlap)
                    # Add any other necessary arguments for inference.py here if needed
                ], stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True) # Capture stdout and stderr

                # Read and print the output in real-time
                while True:
                    output_line = process.stdout.readline()
                    if output_line == '' and process.poll() is not None:
                        break
                    if output_line:
                        print(f"  [inference.py output]: {output_line.strip()}")

                # Check the return code after the process finishes
                return_code = process.wait()
                if return_code != 0:
                    print(f"🔴 ERROR processing {input_file}:")
                    print(f"Command failed with exit code {return_code}")
                    # Any remaining output in stderr/stdout will be printed by the while loop
                else:
                    print(f"✅ Successfully finished processing {input_file}")


            except FileNotFoundError:
                 print(f"🔴 ERROR: 'python' or 'inference.py' not found when processing {input_file}.")
                 print("Please ensure Python is installed and you are in the correct directory where inference.py exists.")
                 print("Continuing with the next file...")
            except KeyboardInterrupt:
                print(f"\n⚠️ Processing of {input_file} interrupted by user.")
                # You might want to add logic here to clean up any partial output files
                break # Exit the loop on interruption
            except Exception as e:
                print(f"🔴 An unexpected error occurred while processing {input_file}: {e}")
                print("Continuing with the next file...")


        print("\nInference process completed (or interrupted).")

    else:
        print("No input files to process after directory check.")
```

```python
#@markdown #Download Processed Files (Session Storage)

import os
from google.colab import files

# Check if the input mode was Session Storage
if 'input_mode' in locals() and input_mode == "Session Storage":
    session_output_dir = "/content/session_output"
    print(f"Checking for processed files in session storage output directory: {session_output_dir}")

    if os.path.isdir(session_output_dir):
        processed_files = [f for f in os.listdir(session_output_dir) if os.path.isfile(os.path.join(session_output_dir, f))]

        if processed_files:
            print(f"\nFound {len(processed_files)} processed files to download:")
            for i, file_name in enumerate(processed_files):
                file_path = os.path.join(session_output_dir, file_name)
                print(f"  {i+1}. Downloading: {file_name}")
                try:
                    files.download(file_path)
                    print(f"    ✅ Download initiated for {file_name}")
                except Exception as e:
                    print(f"    🔴 ERROR initiating download for {file_name}: {e}")
        else:
            print("⚠️ No processed files found in the session storage output directory.")
    else:
        print(f"🔴 ERROR: Session storage output directory not found at {session_output_dir}.")
else:
    print("Download step skipped: Input mode is not 'Session Storage'.")
```
